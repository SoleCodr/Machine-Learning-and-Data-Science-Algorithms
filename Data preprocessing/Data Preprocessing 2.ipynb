###### Outlier treatment

<b>What is an outlier ?</b>
outlier is a commonly used terminology by analysts and data scientists, Outlier is an observation that appears far away and diverges from an overall pattern in a sample.

###### Outlier treatment

<b>Occurence of outliers may be because of</b>
Data entry error, Measurement error, Sampling error. 

<b>It can cause</b>
increase in the error and can reduce the power of statistical tests, lesser prediction accuracies.Mean of the data shifts towards the outlier, standard deviation also differs a lot in data having outliers.

<b>These can be solved by</b>
detecting them using EDD, visualization methods like scatter plot, histogram, box plots. Imputing the outliers can also help.



One method of treating the outliers is <b> capping and flooring</b>. 
- impute all the values above 3*P99 and below 0.3*P1
- impute with values 3*P99 and 0.3*P1

```python
df.info()
```

```python
np.percentile(df.n_hot_rooms, [99])[0]
```

```python
upper_limit=np.percentile(df.n_hot_rooms, [99])[0]
```

```python
df[(df.n_hot_rooms>upper_limit)]
```

```python
df.n_hot_rooms[(df.n_hot_rooms> 3* upper_limit)] = 3* upper_limit
```

```python
df[(df.n_hot_rooms>upper_limit)]
```

```python
lower_limit = np.percentile(df.rainfall, [1])[0]
```

```python
df[(df.rainfall<lower_limit)]
```

```python
df.rainfall[(df.rainfall< 0.3* lower_limit)] = 0.3* lower_limit
```

```python
df[(df.rainfall<lower_limit)]
```

##### Missing value imputation

1. Impute with zero
2. Impute with median/mean/mode
    - for numerical values impute them with mean, median.
    - for categorical data impute them with mode.
3. Segment based imputation
    - if rainfall data of a city is missing we can replace it with mean of the cities belonging to same region maybe.

#### Handling Categorical data

A dummy variable or indicator variable is an artificial variable created to represent an attribute with two or more distinct categories. 

why ?

- regression analysis treats all independent variables in the analysis as numerical. 
- Nominal variables or variables that describe a characterstic using two or more categories are commonplace in regression research, but are not always useable in their categorical form. 
- Dummy coding is a way of incorporating nominal variables into regression analysis. 

How ?

- We can make a separate column, or variable, for each category.
- This new variables can take value 0 or 1 depending on the value of the categorical variable. 

```python
df.head()
```

data has 2 categorical variables waterbody and airport 


```python
df = pd.get_dummies(df)
```
