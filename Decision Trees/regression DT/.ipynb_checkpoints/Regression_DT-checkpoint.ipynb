```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Movie_regression.csv", header=0)
```

```python
df.head()
```

```python
df.info()
```

<b>Missing values treatment<b>

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value=df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

<b> Categorical data </b>

```python
df = pd.get_dummies(df, columns = ['3D_available','Genre'], drop_first = True)
```

```python
df.head()
```

<b>Splitting data</b>

```python
X = df.loc[:, df.columns!='Collection']
type(X)
```

```python
y = df['Collection']
type(y)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2, random_state=0)
```

<b>Training Regression Tree</b>

```python
from sklearn import tree
reg = tree.DecisionTreeRegressor(max_depth = 3)
```

```python
reg.fit(X_train, y_train)
```

<b>Predicting</b>

```python
y_train_pred = reg.predict(X_train)
y_test_pred = reg.predict(X_test)
```

<b>Performance of model</b>

- Mean squarred error: Mean of the square of deviation of predicted values to the actual values.
- r2_score : 1 is perfect fit, 0 is no fit. lies between 0.4 to 0.8 for good models, it is goodness of fit measure.Represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model.

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
r2_score(y_test, y_test_pred)
```

```python
r2_score(y_train, y_train_pred)
```

<b>Plotting decision Tree</b>

```python
from sklearn.tree import export_graphviz      
```

```python
export_graphviz(reg, out_file ='tree.dot')  
```

```python
from IPython.core.display import Image
Image( filename ='tree.png')
```

<b> Controlling Tree growth </b><br/>
<b> Tree Pruning </b>

<b> Maximum number of levels in a tree</b>

```python
regtree1 = tree.DecisionTreeRegressor(max_depth=3)
regtree1.fit(X_train,y_train)
export_graphviz(regtree1, out_file ='tree1.dot', feature_names= X_train.columns, filled = True)  
```

```python
Image(filename ='tree1.png')
```

<b> Minimum observations at the internal node</b>

```python
regtree2 = tree.DecisionTreeRegressor(min_samples_split= 40)
regtree2.fit(X_train,y_train)
export_graphviz(regtree2, out_file ='tree2.dot', feature_names= X_train.columns, filled = True) 
```

```python
Image(filename ='tree2.png')
```

<b> Minimum observations at the leaf node</b>

```python
regtree3 = tree.DecisionTreeRegressor(min_samples_leaf= 25, max_depth = 4 )
regtree3.fit(X_train,y_train)
export_graphviz(regtree3, out_file ='tree3.dot', feature_names= X_train.columns, filled = True) 
```

```python
Image(filename ='tree3.png')
```
