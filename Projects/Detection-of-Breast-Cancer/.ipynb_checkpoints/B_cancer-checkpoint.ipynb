```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df= pd.read_csv("aa.txt", sep = ";")
df.head()
```

```python
df.drop('ID', axis  =1).hist(figsize=(20,8))
```

```python
sum(df.duplicated())
```

```python
df.drop_duplicates()
```

```python
df1 = df.drop('SVM_Classifier',axis=1)
target = df['SVM_Classifier']
x,tx,y,ty = train_test_split(df1,target,test_size=0.2,random_state=0)
```

```python
x.shape,y.shape,tx.shape,ty.shape
```

```python
model = SVC()
model.fit(x,y)

pred = model.predict(tx)
round(accuracy_score(pred,ty)*100,2)
```
