```python
import bs4  #beautifulSoup does the work of extracting the info
```

```python
import requests #does the work of reading the information
```

```python
resp=requests.get('https://www.ducatindia.com/')
```

```python
data=resp.content
```

```python
soup=bs4.BeautifulSoup(data,'html.parser')
```

```python
soup.find('title')
```

```python
soup.find('ul')
```

```python
all_uls=soup.find_all('ul')
```

```python
for ul in all_uls:
    all_lis=ul.findChildren('li')
    for li in all_lis:
        print(li.getText())
```
